```python
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont

# Function to resize images to the same width
def resize_to_same_width(img1, img2, target_width):
    img1 = img1.resize((target_width, int((target_width / img1.width) * img1.height)), Image.Resampling.LANCZOS)
    img2 = img2.resize((target_width, int((target_width / img2.width) * img2.height)), Image.Resampling.LANCZOS)
    return img1, img2

# Open the images
img1 = Image.open(r"C:\Users\Ranga\OneDrive\Pictures\hugging face images\rapido.jpg")
img2 = Image.open(r"C:\Users\Ranga\Downloads\meme_4.PNG")

# Desired width (increase this to make the images larger)
desired_width = 800  # Adjust as needed

# Resize images to the same width
img1, img2 = resize_to_same_width(img1, img2, desired_width)

# Determine the size of the new image
width = max(img1.width, img2.width)  # Max width of the images
height = img1.height + img2.height + 150  # Combined height of both images plus extra space for text

# Create a new blank image
canvas = Image.new('RGB', (width, height), color=(0, 0, 0))

# Paste the first image
canvas.paste(img1, (0, 0))

# Paste the second image below the first one
canvas.paste(img2, (0, img1.height))

# Initialize ImageDraw
draw = ImageDraw.Draw(canvas)

# Prepare the single text element
text = '''That Rapido Anna Earning 3275 per day, 
but me who finished studies in 2023 
still learning "Data Science" and "Gen-AI" .'''
font = ImageFont.truetype(r"C:\Users\Ranga\Downloads\Roboto\Roboto-Medium.ttf", 41)

# Calculate the bounding box for the text
bbox = draw.textbbox((0, 0), text, font=font)
text_width = bbox[2] - bbox[0]
text_height = bbox[3] - bbox[1]

# Calculate the position for the text
text_x = (width - text_width) // 2
text_y = img1.height + img2.height - 595

# Add the text to the canvas
draw.text((text_x, text_y), text=text, fill=(255, 25, 25), font=font, align='center')

# Save the new image
canvas.save(r"C:\Users\Ranga\OneDrive\Pictures\hugging face images\merged_image_with_single_text.jpg")

# Display the new image using matplotlib
plt.imshow(canvas)
plt.axis('on')  # Hide the axis
plt.show()
```
